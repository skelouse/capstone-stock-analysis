```python
import pandas as pd
import plotly
```

```python
def get_symbol_frequency(df, sym, freq=None):
    perf_min = df.reset_index()['date'].min()
    perf_max = df.reset_index()['date'].max()
    
    if freq:
        weekly_cut = pd.date_range(perf_min, perf_max, freq=freq)
        q=df.query("sym == '%s'"%sym).reset_index().set_index('date', drop=True).reindex(weekly_cut)
    else:
        q=df.query("sym == '%s'"%sym).reset_index().set_index('date', drop=True)
    return q.dropna()
```

```python
df_performance = pd.read_pickle('./data/norm.pkl')
df_penny_performance = pd.read_pickle('./data/penny.pkl')
df_perc = pd.read_pickle('./data/perc.pkl')
```

```python
get_symbol_frequency(df_performance, 'AAPL', freq=None).head()
```

```python
import plotly.express as px

fig = px.line(get_symbol_frequency(df_performance, 'AAPL', freq=None).reset_index(), x='date', y='price')
fig.show()
```
